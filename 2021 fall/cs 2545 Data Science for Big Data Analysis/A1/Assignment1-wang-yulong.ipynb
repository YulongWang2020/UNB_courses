**Q2.2**  Using the dataframe from above, group the countries by *IncomeGroup* column and show the average GDP for each income group.

**Q3.0** Download the *capitals.csv*  and *major_cities.csv* files from D2L in the current folder. Load them into two dataframes and show first 5 rows. The major cities are some of the cities with relatively large population. The capitals are capitals of the countries.


```python
# Load capitals.csv

capitals = pd.read_csv('capitals.csv',sep=',')
capitals.head()
```

```python
# Load major_cities.csv

major_cities = pd.read_csv('major_cities.csv',sep=',')
major_cities.head()
```

**Q3.1** For each country that has a Megacity (i.e. city with population more than 10 million), show the name of country and the number of Megacities it has. Note, the countries should be sorted in descending order based on the number of megacities they have.

```python
Megacity = major_cities[major_cities['CityPopulation']>10e6]
Megacity[['Country','City']].groupby('Country').count().sort_values('City',ascending=False)
```

**Q3.2** For the countries each of whose capital is a major city, show the name of the top 7 capitals (ordered by thier capital city population) and the population of its capital, along with the correspinding country name.

```python
capital_major_city = pd.merge(capitals,major_cities,left_on='CapitalCity',right_on='City').sort_values('CityPopulation',ascending=False)
capital_major_city = capital_major_city.rename(columns={'Country_x':'Country Name'})
capital_major_city[['CapitalCity','CityPopulation','Country Name']].head(7)
```

**Q4.1** From question Q3.2, plot the name of the top 7 capitals (i.e. by capital city population) in a map of the world.  Use the Basemap module for this.

```python
capital_major_city
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


city_name = capital_major_city['CapitalCity'].values[:7]
population = capital_major_city['CityPopulation'].values[:7]


lats = capital_major_city['Latitude'].values[:7]
lons = capital_major_city['Longitude'].values[:7]

fig = plt.gcf()
fig.set_size_inches(18, 12)

m = Basemap(projection='merc', 
            llcrnrlat=-80, urcrnrlat=80, 
            llcrnrlon=-180, urcrnrlon=180,
            lat_ts=20,
            resolution='c')

m.bluemarble(scale=0.2)
m.drawcoastlines(color='white', linewidth=0.2)

x, y = m(lons, lats)  # transform coordinates
plt.scatter(x, y, 30, marker='o', color='Red')
for i in range(7):
    plt.text(x=x[i]+0.01*x[i], y=y[i]+0.02*y[i], s=city_name[i], fontdict=dict(color='white',size=15))
plt.show()
```
